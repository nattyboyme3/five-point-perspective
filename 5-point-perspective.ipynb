```python
import drawSvg as draw
import numpy as np

r = 500
d = draw.Drawing(2*r+50, 2*r+50, origin='center', displayInline=False)
# Draw a circle
d.append(draw.Circle(0, 0, r,
            fill='transparent', stroke_width=2, stroke='black'))
d.append(draw.Lines(r,0,-r,0, stroke_width=1, stroke='red'))
d.append(draw.Lines(0,r,0,-r, stroke_width=1, stroke='red'))

i =5
increment = i 
while i < 90:
    g = i
    x = 1/np.sin(np.deg2rad(g))
    t = 1/np.tan(np.deg2rad(g))
    h = t * r
    
    cds = r * r
    bd = x*r
    bds = bd*bd
    arc_radius = (bds + cds)/(2 * bd)
    offset = arc_radius - bd
    
    # do "horizontal" lines
    d.append(draw.ArcLine(0,offset,arc_radius,90-(g*2),90+(g*2),
                stroke='red', stroke_width=1, fill='transparent'))
    d.append(draw.ArcLine(0,-offset,arc_radius,270-(g*2),270+(g*2),
                stroke='red', stroke_width=1, fill='transparent'))
    
    # do "vertical" lines
    d.append(draw.ArcLine(offset,0,arc_radius,0-(g*2),0+(g*2),
                stroke='red', stroke_width=1, fill='transparent'))
    d.append(draw.ArcLine(-offset,0,arc_radius,180-(g*2),180+(g*2),
                stroke='red', stroke_width=1, fill='transparent'))
    
    i = i + increment
d.append(draw.Circle(0,0,r+100, fill='transparent', stroke_width=200, stroke='white'))
d  # Display as SVG
```

```python
import drawSvg as draw
import numpy as np

r = 200
d = draw.Drawing(2*r+50, 2*r+50, origin='center', displayInline=False)
# Draw a circle
d.append(draw.Circle(0, 0, r,
            fill='transparent', stroke_width=2, stroke='black'))
d.append(draw.Lines(r,0,-r,0, stroke_width=1, stroke='red'))
d.append(draw.Lines(0,r,0,-r, stroke_width=1, stroke='red'))

lines = 20
r_div = r/lines
for i in range (1, lines):
    d.append(draw.Ellipse(0,0,r,i*r_div, fill='transparent', stroke_width=1, stroke='red'))
    d.append(draw.Ellipse(0,0,i*r_div,r, fill='transparent', stroke_width=1, stroke='red'))

d  # Display as SVG
d.exportSvg()
```
